```python
# TODO: Create dataframe of airline_iata, number of incidents, number of routes
incidents_df['Airline'] = incidents_df['Airline'].str.lower()
airlines_df['AirlineName'] = airlines_df['AirlineName'].str.lower()
m1 = airlines_df.merge(routes_df,left_on = 'IATA', right_on = 'IATAICAOAirline', how = 'inner')
m1 = m1.groupby(['IATA','AirlineName'],as_index=False).count()
m1 = m1.rename(index=str, columns={"AirlineID_x": "routes_count"})[['IATA','AirlineName','routes_count']]
m2 = m1.merge(incidents_df, left_on = 'AirlineName', right_on = 'Airline', how = 'inner')
m2 = m2.groupby(['IATA','routes_count','AirlineName'],as_index=False).count().rename(index=str, columns={"Date": "incidents_count"})[['IATA','routes_count','incidents_count','AirlineName']]
m2
```

```python
res = m2[m2['incidents_count']>2]
res
```

```python
import numpy as np
import matplotlib.pyplot as plt

i = res['routes_count'].values
j= res['incidents_count'].values
z = res['IATA'].values

fig, ax = plt.subplots()
ax.scatter(i,j, marker='o')

for label, x, y in zip(z, i, j):
    plt.annotate(label, xy=(x, y), xytext=(-1, 1), textcoords='offset points')

plt.xlabel('Number of routes')
plt.ylabel('Number of incidents')
plt.show()
```

## Tasks for Step 5

```python
# TODO: get random sample of flights from SQL
flights_s_df = pd.read_sql_query('SELECT * FROM flights ORDER BY RANDOM() LIMIT 1000', con=engine)
```

```python
# TODO: create nested map of Jaccard distance
def jaccard_distance(A,B):
    return len(A.intersection(B))/len(A.union(B))
```

## Step 5.1

```python
# find non-numeric columns of routes and flights
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
routes_nonnum = routes_df.select_dtypes(exclude=numerics)
flights_s_nonnum = flights_s_df.select_dtypes(exclude=numerics)
```

```python
# TODO: output nested map
dict = {}
for r in routes_nonnum.columns:
    dict[r] = {}
    for p in flights_s_nonnum.columns:
        routes_col = routes_nonnum[r].drop_duplicates()
        flights_col = flights_s_nonnum[p].drop_duplicates()
        d = jaccard_distance(set(routes_col),set(flights_col))
        if d > 0:
            if r not in dict:
                dict[r] = {}
            dict[r][p] = d
dict
```
